```python
import HP

def get_sent(pid):

    f = open(HP.data_directory + pid + '.txt')

    waiting_for_new_sentence_flag = True
    ans_ls = []
    ans = ''

    for line in f:
        strip_line = line.strip()
        if len(strip_line) == 0:
            waiting_for_new_sentence_flag = True
            ans_ls.append(ans)
            ans = ''
            continue
        elif waiting_for_new_sentence_flag:
            waiting_for_new_sentence_flag = False
            ans += strip_line + ' '
            current_word_ind = 0
        else:  
            if current_word_ind < HP.n_max_word_num:
                ans += strip_line + ' '
                current_word_ind += 1

    f.close()
    
    return ans_ls

def get_scores_dict(type_data):

    scores_dict = {}
    waiting_for_new_sentence_flag = True
    pid = ''
    y = -1
    scores = []

    f = open('results/{}_scores.txt'.format(type_data))
    for line in f:
        strip_line = line.strip()
        if len(strip_line) == 0:
            waiting_for_new_sentence_flag = True
            scores_dict[pid] = [y, scores]
            scores = []
        elif waiting_for_new_sentence_flag:
            waiting_for_new_sentence_flag = False
            pid,y = strip_line.split(', ')
        else:
            scores.append(float(strip_line))
    
    return scores_dict

scores_dict_train = get_scores_dict('train')
scores_dict_dev = get_scores_dict('dev')
scores_dict_test = get_scores_dict('test')

def find_topn(ls):
    
    ls = list(ls)
    ans = []
    n = len(ls)
    
    while n != 0:
        idx = ls.index(max(ls))
        ans.append(idx)
        ls[idx] = -1
        n -= 1
        
    return ans

def printSent(type_data, pid):

    if type_data =='train':
        scores_dict = scores_dict_train
    if type_data == 'dev':
        scores_dict = scores_dict_dev
    if type_data == 'test':
        scores_dict = scores_dict_test
    
    sent_ls = get_sent(pid)
    y, scores = scores_dict[pid]
    
    topn = find_topn(scores)
        
    print(pid)
    print('prediction: '+ y+ '\n')
        
    for t in topn:
        print(str(scores[t]) + ' --- ' + sent_ls[t])
        
f = open('results/test_output.txt')

true_positive = []
true_positive_prediction = []
true_negative = []
true_negative_prediction = []
false_positive = []
false_negative = []

for line in f:
    if line[0]=='-' or line[-2]=='n':
        continue
    pid, true, prediction = line.split(',')
    if int(true)==1 and float(prediction)>=0.5:
        true_positive.append(pid)
        true_positive_prediction.append(float(prediction))
    elif int(true)==0 and float(prediction)<0.5:
        true_negative.append(pid)
        true_negative_prediction.append(float(prediction))
    elif int(true)==0 and float(prediction)>=0.5:
        false_positive.append(pid)
    elif int(true)==1 and float(prediction)<0.5:
        false_negative.append(pid)
        
print('true positive: ', len(true_positive))
print('true negative: ', len(true_negative))
print('false positive: ', len(false_positive))
print('false negative: ', len(false_negative))
```

### aggregate vocab

```python
import collections
import pandas as pd
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```python
vocab = []

for pid in true_positive:
    
    sent_ls = get_sent(pid)
    y, scores = scores_dict_test[pid]
    
    top10 = find_topn(scores)[:10]
    
    for t in top10:
        vocab.extend(sent_ls[t].strip().split(' ')[1:])
        
pd.DataFrame(collections.Counter(vocab).most_common()).head(50)
```

```python
vocab_clean = [v for v in vocab if v not in stopwords]
vocab_clean = [v for v in vocab_clean if v.isalpha()]

pd.DataFrame(collections.Counter(vocab_clean).most_common()).head(50)
```
